```python
## PHTHON STRING                   PHTHON STRING                      PHTHON STRING                  PHTHON STRING            PHTHON STRING        PHTHON STRING 

# Get subset of a string. You use [:]. In the example one below, you get the first character in the string starting at position 2 through 5. Note that the character in 5 string will be excluded.

# Example 1
b = "Hello, World!"
print(b[2:5])


# Get the characters from position 5 , starting the count from the end of the string: Note that the character in -2 (postion 2 from the back) string will be excluded. The counting from back start 
# at 1 not zero.

# Example 2
b = "Hello, World!"
print(b[-5:-2])

# To get the length of a string, use the len() function.

a = "Hello, World!"
print(len(a))


###### STRING METHODS :   Python has a set of built-in methods that you can use on strings ######

"""
The strip() method removes any whitespace from the beginning or the end
The lower() method returns the string in lower case
The upper() method returns the string in upper case:
The replace() method replaces a string with another string:
The split() method splits the string into substrings if it finds instances of the separator you provided.
"""

# strip() example
a = "       Hello, World!      "
print(a.strip()) # returns "Hello, World!"

# lower() example
a = "Hello, World!"
print(a.lower())

# upper() example
a = "Hello, World!"
print(a.upper())

# get rid of whitespace and then turn to lower case
a = "       Hello, World No WHITESPACE IN LOWERCASE      "
a = a.strip()
print(a.lower())

# replace example: replace H with J

a = "Hello, World!"
print(a.replace("H", "J"))

# split() example
a = "Hello, World!. The names are John, Michael, Ade , Love and Funmi"
print(a.split(",")) # returns ['Hello', ' World!']


####### CHECK STRING : To check if a certain phrase or character is present in a string, we can use the keywords in or not in.

# Example : Check if the phrase "ain" is present in the following text:
txt = "The rain in Spain stays mainly in the plain"
x = "ain" in txt
print(x)

# Example : Check if the phrase "ain" is not present in the following text:
txt = "The rain in Spain stays mainly in the plain"
x = "ain"  not in txt
print(x)

# concatenate a and b and separate by spsce.

a = "Hello"
b = "pretty"
c = a + " , " + b
print(c)



```
