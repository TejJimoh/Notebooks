```python
"""
What is Standard Deviation?
Standard deviation is a number that describes how spread out the values are.
A low standard deviation means that most of the numbers are close to the mean (average) value.
A high standard deviation means that the values are spread out over a wider range.
The Standard Deviation and Variance are terms that are often used in Machine Learning, 
so it is important to understand how to get them, and the concept behind them.
"""

# The NumPy module has a method to calculate the standard deviation:
# Use the NumPy std() method to find the standard deviation:

import numpy
speed = [86,87,88,86,87,85,86] # The numbers in the list are very close to one another. Hence, standard deviation will be low.
x = numpy.std(speed)  ##### REMEMBER numpy.std() to calculate the standard deviation #####
print(x) 

# The numbers in the list below are well spread out. Hence, standard deviation will be high.

import numpy
speed = [32,111,138,28,59,77,97]
x = numpy.std(speed)
print(x) 


#### Variance is another number that indicates how spread out the values are.
# A low variance means that most of the numbers are close to the mean (average) value.
# A high variance means that the values are spread out over a wider range.

# Use the NumPy var() method to find the variance:
# The numbers in the list below are well spread out. Hence, variance will be high.
import numpy
speed = [32,111,138,28,59,77,97]
x = numpy.var(speed)  ##### REMEMBER numpy.var() to calculate the variance #####
print(x) 

###### REMEMBER numpy.std() , numpy.var()  
```

```python
""" Percentiles are used in statistics to give you a number that describes the value that a given percent of the values are lower than.
Example: Let's say we have an array of the ages of all the people that lives in a street.
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31] 
What is the 75. percentile? The answer is 43, meaning that 75% of the people are 43 or younger.
"""


# Use the NumPy percentile() method to find the percentiles: 75% of ages are 43 years or younger

import numpy
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
x = numpy.percentile(ages, 75)   ##### REMEMBER numpy.percentile() to calculate percentile #####
print(x)

# What is the age that 90% of the people are younger than?
import numpy
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
x = numpy.percentile(ages, 90) # It takes 2 arguments namely the variable and percentage.
print(x) 

 ##### REMEMBER numpy.percentile() to calculate percentile #####
```

```python
## How Can we Get Big Data Sets? We can use python uniform data distribution.
## To create big data sets for testing, we use the Python module NumPy, which comes with a number of methods to create random data sets, of any size.
## Create an array containing 250 random floats between 0 and 5:

import numpy
x = numpy.random.uniform(0.0, 5.0, 250) ##### REMEMBER numpy.random.uniform() for uniform data distribution #####
print(x) 


# Histogram
# To visualize the data set we can draw a histogram with the data we collected.
# We will use the Python module Matplotlib to draw a histogram:
# Example: Draw a histogram:
# The import statement calls the matplotlib module.The next step is to import the pyplot plotting framework that is included in the matplotlib package:

import numpy
import matplotlib.pyplot as plt 
x = numpy.random.uniform(0.0, 5.0, 250) ## Create an array containing 250 random floats between 0 and 5:
plt.hist(x, 5)  ## call the histogram method hist() to display 5 bars using pyplot framework of matplotlib
plt.show()  ## plot the graph

##### REMEMBER : numpy.random.uniform(), matplotlib.pyplot as plt , plt.hist(), plt.show()

## In a nutshell, you need numpy.random.uniform() to generate random uniform numbers. It takes 3 parameters. 
## The generated numbers will be between the first 2 parameters while the third parameter is the total number to be generated.
## matplotlib.pyplot is require for plotting purposes. plt is just an alias.
## plt.hist() simply means the type of graph to be plotted should be an histogram.
## plt.show() needed to display the plot.
```

```python
## Normal Data Distribution
## In this chapter we will learn how to create an array where the values are concentrated around a given value.
## This kind of data distribution is known as the normal data distribution, or the Gaussian data distribution.

## A typical normal data distribution:

import numpy
import matplotlib.pyplot as plt
x = numpy.random.normal(5.0, 1.0, 100000) # set mean value to 5, standard deviation to 1 and generate 100000 values. use numpy.random.normal
plt.hist(x, 100)
plt.show() 

# We use the array from the numpy.random.normal() method, with 100000 values,  to draw a histogram with 100 bars.
# We specify that the mean value is 5.0, and the standard deviation is 1.0.
# Meaning that the values should be concentrated around 5.0, and rarely further away than 1.0 from the mean.
# And as you can see from the histogram, most values are between 4.0 and 6.0, with a top at approximately 5.0.

##### REMEMBER : numpy.random.normal(): The parameters are mean, standard deviation and total numbers to be generated.
```

```python
#### LINEAR REGRESSION            LINEAR REGRESSION            LINEAR REGRESSION            LINEAR REGRESSION            LINEAR REGRESSION

"""
Regression:
The term regression is used when you try to find the relationship between variables.
In Machine Learning, and in statistical modeling, that relationship is used to predict the outcome of future events.

Linear Regression
Linear regression uses the relationship between the data-points to draw a straight line through them.
This line can be used to predict future values.

"""

# In the example below, the x-axis represents age, and the y-axis represents speed. 
# We have registered the age and speed of 13 cars as they were passing a tollbooth. 
# Let us see if the data we collected could be used in a linear regression:

import matplotlib.pyplot as plt     # First step is to import your modules
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]      # Create the arrays that represents the values of the x and y axis:
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)  # Draw the original scatter plot:
plt.plot(x, mymodel)  # Draw the line of linear regression:
plt.show()  # Display the diagram
# print(slope)
# print(intercept)

# Let's explain above linear regression in details

"""
STEP 1: import your modules

import matplotlib.pyplot as plt 
from scipy import stats

STEP 2:  Create the arrays that represents the values of the x and y axis:

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]      
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

STEP 3: Execute a method that returns some important key values of Linear Regression:

slope, intercept, r, p, std_err = stats.linregress(x, y).
The function scipy.stats.linregress(x,y) performs linear regression of y on x. It returns 5 values namely : the slope, intercept,
correlation, p-value against the null hypothesis that the slope is zero, and lastly standard_error.

STEP 4: Create a function that uses the slope and intercept values to return a new value. This new value represents  y-axis :

def myfunc(x):
  return slope * x + intercept

This function returns y as you pass in x values using y = mx + c. where m is the slope or gradient (dy/dx) of the line and 
c is the y-intercept (where the line crosses the y-axis).

You can get the 5 values by using : print(slope), print(intercept) etc etc 

STEP 5 : Uses Python map() function to apply a function on all the elements of specified iterable and return map object. 

mymodel = list(map(myfunc, x)) 

we basically pass in myfunc() function ( with y values) and x iterators to map and convert the map to list. 

"""
```

```python
"""

R-Squared : Relationship or correlation between x and y.
It is important to know how the relationship between the values of the x-axis and the values of the y-axis is, if there are no relationship the linear 
regression can not be used to predict anything.
The relationship is measured with a value called the r-squared.
The r-squared value ranges from -1 to 1, where 0 means no relationship, and 1, or -1, means 100% related.
Python and the Scipy module will compute this value for you, all you have to do is feed it with the x and y values:

"""
# How well does my data fit in a linear regression?
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
## Note: The result -076 shows that there is a relationship, not perfect, but it indicates that we could use linear regression in future predictions.


###  Predict Future Values         Predict Future Values          Predict Future Values       Predict Future Values

# Now we can use the information we have gathered to predict future values.
# Example: Let us try to predict the speed of a 10 years old car.
# To do so, we need the same myfunc() function from the example above: Note that myfunc() returns y which is the speed of the car.
# You simply pass the predictive value of 10 to the function

## Predict the speed of a 10 years old car: You can see that the printed value correspond to the value of 85.6 on the graph.
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

speed = myfunc(10)
print(speed) 
```
